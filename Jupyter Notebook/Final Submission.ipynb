# RICE
Same Features as above plus some Specific Features:
## 1.RICE EXPORT IN THAT YEAR
## 2.RICE EXPORT VALUE IN THAT YEAR

We are having a final DataSet as Follows:

```python
import pandas as pd
data=pd.read_csv('RiceFinal.csv')
```

```python
data.set_index('Unnamed: 0',inplace=True)
```

```python
data
```

### Different Scale of Values of Data
We can see that In Rice Exports and Rice Exports Values The Values Are Very Large in Scale as Comparison to other Values,
Therefore, we will do Standard Scaling Of Data to use Linear Regression(It's Results are highly senstitve to Scaling Of Data),but let's also pop out Target Value(Price) 

```python
target = data.pop('Rice.Price(Target)')
from sklearn.preprocessing import StandardScaler
data['Rice Export Quantity']=data['Rice Export Quantity']/1000000
scaler = StandardScaler()
scaleddata=scaler.fit_transform(data.values)
```

Exports and Rainfall play a proportional Role in Prices of Rice, So we Will Try to Train a LinearRegression Model To Satisfy Our Needs

```python
from sklearn.linear_model import LinearRegression
LinReg=LinearRegression()
```

We will use Train_test_Split Method to Train And Test Our Model,

```python
from sklearn.model_selection import train_test_split
A,B,C,D = train_test_split(scaleddata,target,test_size=0.3)
```

```python
LinReg.fit(A,C)
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(LinReg.predict(B),D)
```

# JOWAR

Similarly We have A Dataset For It.

```python
datajowar=pd.read_csv('jowarfinal1.csv')
from sklearn.preprocessing import StandardScaler
jowscale=StandardScaler()
```

```python
datajowar.set_index('Unnamed: 0',inplace=True)
```

```python
datajowar
```

```python
target=datajowar.pop('Jowar-Price(Target)')
```

```python
from sklearn.ensemble import RandomForestClassifier
jowarRand = RandomForestClassifier()
e=target.values
d1=jowscale.fit_transform(datajowar.values)
```

```python
import numpy as np
m=np.where(e>=150,(np.where(e>=220,'A+','A')),'B')
```

```python
A,B,C,D=train_test_split(d1,m,test_size=0.3)
jowarRand.fit(A,C)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(jowarRand.predict(B),D)
```
